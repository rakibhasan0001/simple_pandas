```python
import pandas as pd
drinks=pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/drinks.csv')
drinks.head()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.groupby('continent'=='Africa').beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa']
```

```python
drinks.continent=='Africa'
```

```python
# item[bool] give an total list
drinks.groupby('continent').beer_servings.max()
```

```python
# important
# multiple aggrigation function 
drinks.groupby('continent').beer_servings.agg(['mean','max','count','min'])

#this is most powerfull tool so use it
```

```python
drinks.groupby('continent').mean()
```

```python
%matplotlib inline

drinks.groupby('continent').mean().plot(kind='bar')
```
