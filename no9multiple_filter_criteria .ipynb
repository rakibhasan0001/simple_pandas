```python
import pandas as pd
movies=pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv')
movies.head()
```

```python
# & =and |= or
movies[(movies.duration >= 200) & (movies.genre == 'Crime')]
# give brakect to saparate two value and it is one type of evaluation
```

```python
movies[(movies.genre=='Crime') | (movies.genre=='Drama') | (movies.genre=='Action')]
```

# shortcut

```python
#important isin method
 
    
movies[movies.genre.isin(['Action','Drama','Crime'])]
#isin method take two positional argument isin([list]) 1.tupe 2.list
```

```python
import pandas as pd
df = pd.DataFrame({'num_legs': [2, 4], 'num_wings': [2, 0]},
                  index=['falcon', 'dog'])
df
```

```python
# find 2 and 0 in eatch row
df.isin([0,2])
```

```python
df = pd.DataFrame({'num_legs': [2, 4], 'num_wings': [2, 0]},
                  index=['falcon', 'dog'])
print(df)
# you can see eatch column separately
df.isin({'num_wings':[2,0]})#please use dictionary {}
```

```python
other = pd.DataFrame({'num_legs': [8, 2], 'num_wings': [0, 2]},
                     index=['spider', 'falcon'])
df.isin(other)
```
