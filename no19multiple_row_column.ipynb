```python
import pandas as pd
ufo=pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv')
ufo.head(3)
```

```python
# ufo.loc[row:column]
ufo.loc[0,:]
```

```python
ufo.loc[[1,2,4],:]
```

```python
ufo.loc[1:3,:]
```

```python
ufo.loc[1:4,['City','State']]
```

```python
ufo.loc[1:4,'City':'State']
```

```python
ufo[ufo.City == 'Oakland']
```

```python
ufo.loc[ufo.City == 'Oakland',:]
```
